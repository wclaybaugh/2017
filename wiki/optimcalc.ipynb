## The Basic Idea of gradient Descent
The gradient of a smooth, single-output function can be found. The gradient takes in any point in the function's domain and returns a vector pointing in the direction where the function growing fastest. In the opposite direction the function is decreasing fastest. The gradient is extremely local and extrapolating in the suggested direction relies on extra details of the function. See the notebook on calculus for more.

In gradient descent, we follow the gradient's advice and, from some starting point, take a step in the direction suggested by the gradient, find a new gradient, and repeat. In fact, we'll listen to the gradient even more and decide that each step is some fixed size, multiplied by the magnitude of the gradient.

The direction does not matter in 1d (the only options are left or right), but the idea in 1d is the same: you move by the amount of your derivative times a step size. Too small a step, and you take for ever  to go down the hill. To much a step and you overshoot, oscillate around, and then finally make it.

Step size is important.

```python
#adapted from http://tillbergmann.com/blog/python-gradient-descent.html

def grad_des1d(f, fprime, step=0.01, maxsteps=0, precision=0.001, start=25):
    mins = []
    costs = []
    old=0
    current = start
    mins.append(start)
    counter = 0
    while abs(current - old) > precision:
        old = current 
        gradient = fprime(old) 
        move = gradient * step
        current = old - move
        mins.append(current)
        if maxsteps:
            counter+=1
            if counter == maxsteps:
                break
 
    return mins
```

```python
def animator(f, xvals, xdatafn, ydatafn, frameskip, ax, line):
    
    fvals=[f(x) for x in xvals]
    ax.plot(xvals, fvals, lw=3)
    line, = ax.plot([], [], "D", markersize=12)

    def init():
        line.set_data([], [])
        return line,

    def animate(i):
        x_n = xdatafn(i, frameskip)
        y_n = ydatafn(i)
        line.set_data(x_n, y_n)
        return line,
    return init, animate
```

```python
fmins = grad_des1d(f, lambda x: 2*x - 6)
```

```python
fig = plt.figure(figsize=(10, 6))
ax = plt.axes()
ax.plot(xlist,flist, linewidth=4 )
line, = ax.plot([], [], "D", markersize=12)
fskip=10
xdatafn = lambda i, fskip: fmins[0::fskip][i]
ydatafn = lambda i: f(xdatafn(i, fskip))
i, a = animator(f, xlist, xdatafn, ydatafn, fskip, ax, line)
anim = animation.FuncAnimation(fig, a, init_func=i,
                        frames=len(fmins[0::fskip]), interval=200)
anim.save('images/1dgd.mp4')
anim
```
